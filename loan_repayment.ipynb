```python
# Create a scatter plot of customer Age vs Credit
import matplotlib.pyplot as plt
plt.scatter(df['Customer_Age'], df['Credit_Score'], c=df['Can_Repay'])
```

```python
# Select the first two columns as features(X) and the last column as the target (Y)
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
# Split the data into traning and test sets with a test size of 20 sample %
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=20)
```

```python
# Apply standard scaling to the traning data to normalize feature
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
# Initialize a logistic Regression classifier
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(x_train, y_train) 
```

```python
# Make predictions on the test data
y_pred = clf.predict(x_test)
y_pred
```

```python
# Calculae and return the accuray score by comparing the true labels with the predicted labels 
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Plot the decision redions of the classifier using the traning data
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train, y_train.values, clf=clf, legend=2)
```
